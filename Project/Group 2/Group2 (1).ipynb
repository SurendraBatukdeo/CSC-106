```python
import csv 
import pandas as pd
import matplotlib.pyplot as plt
#income = int(input("What is your monthly income after taxes?"))
income = 80000

with open('budget-data.csv', 'r') as file:
    reader = csv.DictReader(file)
    data_dict = []
    for row in reader:
        data_dict.append(row)
    for x in data_dict:
        x['Amount'] =  float(x['Amount']) / 100
        x['Amount'] = x['Amount'] * income

# Create dataframe
dataframe = pd.DataFrame(data_dict)
dataframe

# Create Pie
plt.figure(figsize=(8, 8))
plt.pie(dataframe['Amount'], labels=dataframe['Type'], autopct='%1.1f%%', startangle=80, colors=plt.cm.Paired.colors)
plt.title('Expenditure Distribution')
plt.show()
    
```
